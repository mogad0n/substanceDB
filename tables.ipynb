```python
import pandas as pd
df = pd.read_json('./sources/merge-psychonautwiki-tripist-data/pandas_substance_data.json')
df = df.set_index('name')
df = df.rename_axis('name', axis='columns')
df.head()
```

```python
substance_roas = {}
for name in df.index:
    sort = {}
    
    for roa in df.roas[name]:
        key = roa['name']
        sort['dosages'] = {}
        sort['durations'] = {}

        if 'dosage' in roa.keys() and roa['dosage'] != None:
            dosage = roa['dosage']
            sort['dosages'][key] = {}
            for dose in dosage:
                sort['dosages'][key][dose['name']] = dose['value']  
                if 'note' in dose.keys():
                    sort['dose_note'] = dose['note']       
                else:
                    sort['dose_note'] = ''

        if 'duration' in roa.keys() and roa['duration'] != None:
            duration = roa['duration']
            sort['durations'][key] = {}
            for duration in duration:
                sort['durations'][key][duration['name']] = duration['value']
    substance_roas[name] = sort
substance_roas
```

```python
pd.DataFrame.from_dict(substance_roas, orient='index')
```
